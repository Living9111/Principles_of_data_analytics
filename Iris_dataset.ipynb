# Task 1 Source the Data Set

## Importing required Libraries for data set

```python
# Import necessary libraries for Data set creation
import pandas as pd
from sklearn.datasets import load_iris
import numpy as np

# Import necessary libraries for Data set visualization
import matplotlib.pyplot as plt
```

## ABOUT DATASET

### Load Iris data set consists of 3 different types of irisesâ€™ (Setosa, Versicolour, and Virginica) petal and sepal length, stored in a 150x4 

### The rows being the samples and the columns being: Sepal Length, Sepal Width, Petal Length and Petal Width.

```python
# Load the iris dataset
data = load_iris()
```

```python
# Convert the feature data to a DataFrame
target_df = pd.DataFrame(data['data'], columns=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'])

```

# Task 5: Investigate Relationships

#### Choose any two features from the data set and create a scatter plot of them

```python
# Choose two features for the scatter plot
feature_x = 'sepal length (cm)'
feature_y = 'sepal width (cm)'

# Create a scatter plot of the two features
plt.figure(figsize=(10, 6))  # Set the figure size
plt.scatter(target_df[feature_x], target_df[feature_y], c='Blue', alpha=0.5)  # Scatter plot

# Set the title 
plt.title(f"Scatter Plot of {feature_x} vs {feature_y}", fontsize=16)  # Add a title

# Add labels to the axes
plt.xlabel(feature_x, fontsize=14)  # Add x-axis label
plt.ylabel(feature_y, fontsize=14)  # Add y-axis label


plt.show()  # Show the plot
   
```

```python
# Color-code the three different classes of the scatter plot points

plt.figure(figsize=(8, 6))
for target, color, label in zip([0, 1, 2], ['red', 'blue', 'green'], data['target_names']):
    plt.scatter(
        target_df[feature_x][data['target'] == target],
        target_df[feature_y][data['target'] == target],
        color=color,
        label=label,
        alpha=0.7
    )

# Add labels, title, and legend
plt.xlabel(feature_x)
plt.ylabel(feature_y)
plt.title(f"Scatter Plot of {feature_x} vs {feature_y}")
plt.legend(title="Classes")
plt.grid(alpha=0.3)
plt.show()
```

# Task 6: Analyze Relationship

```python
# Use numpy.polyfit to add a regression line to the scatter plot from Task 5.

# Fit a linear regression line to the data
[slope, intercept]= np.polyfit(target_df[feature_x], target_df[feature_y], 1)


# Create a range of x values for the regression line
x = np.linspace(target_df[feature_x].min(), target_df[feature_x].max(), 100)

# Calculate the corresponding y values using the slope and intercept
y = slope * x + intercept

# Create a scatter plot of the two features
plt.figure(figsize=(10, 6))
plt.scatter(target_df[feature_x], target_df[feature_y], c='Blue', alpha=0.5)  # Scatter plot

# Add the regression line
plt.plot(x, y, color='red', label='Regression Line')  # Regression line



# Add labels, title, and legend
plt.xlabel(feature_x, fontsize=14)  # Add x-axis label
plt.ylabel(feature_y, fontsize=14)  # Add y-axis label
plt.title(f"Scatter Plot of {feature_x} vs {feature_y} with Regression Line", fontsize=16)  # Add a title
plt.legend(title="Classes")
```

# Task 8: Compute Correlations

```python
# Calculate the correlation coefficients between the features
correlation_coeff= target_df.corr()

# Display the correlation matrix
print("Correlation coefficient:")
print(correlation_coeff)


# Create a heatmap of the correlation coefficient using matplotlib
plt.figure(figsize=(10, 8))
plt.imshow(correlation_coeff, cmap='coolwarm', interpolation='none')

plt.colorbar()
plt.xticks(range(len(correlation_coeff)), correlation_coeff.columns, rotation=90)
plt.yticks(range(len(correlation_coeff)), correlation_coeff.columns)

# Set the title of the heatmap

plt.title('Correlation Heatmap of Iris Dataset Features')

# Add text annotations to the heatmap
labels = target_df.columns
for x in range(len(labels)):
    for y in range(len(labels)):
        plt.text(y, x, round(correlation_coeff.iloc[x, y], 2), ha='center', va='center', color='white')
```

# Task 9: Fit a Simple Linear Regression

```python
# Fit a linear regression line to the data
[slope, intercept]= np.polyfit(target_df[feature_x], target_df[feature_y], 1)


# Create a range of x values for the regression line
x = np.linspace(target_df[feature_x].min(), target_df[feature_x].max(), 100)

# Calculate the corresponding y values using the slope and intercept
y = slope * x + intercept

# Create a scatter plot of the two features
plt.figure(figsize=(10, 6))
plt.scatter(target_df[feature_x], target_df[feature_y], c='Blue', alpha=0.5)  # Scatter plot

# Add the regression line
plt.plot(x, y, color='red', label='Regression Line')  # Regression line


# Add labels, title, and legend
plt.xlabel(feature_x, fontsize=14)  # Add x-axis label
plt.ylabel(feature_y, fontsize=14)  # Add y-axis label
plt.title(f"Scatter Plot of {feature_x} vs {feature_y} with Regression Line and Annote with Coefficient", fontsize=16)  # Add a title
plt.legend(title="Classes")

#annotate it with the 
plt.annotate(f"y = {slope:.2f}x + {intercept:.2f}", xy=(0.5, 0.5), xycoords='axes fraction', fontsize=12, color='red')


# calculate the coefficient of determination
r_squared = np.corrcoef(target_df[feature_x], target_df[feature_y])[0, 1] ** 2
print(f"Coefficient of Determination (R^2): {r_squared:.2f}")
# Show the plot
plt.show()
```
